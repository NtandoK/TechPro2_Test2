```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('settle2.csv')
```

```python
toiletsTotal=data['toilets_total']
toiletsFullFlush=data['toilets_full_flush']
```

```python
plt.scatter(toiletsTotal, toiletsFullFlush)
plt.title('Toilets Total vs Toilets full flush')
plt.xlabel("Toilets Total")
plt.ylabel("Toilets Full Flush")
plt.show()
```

```python
area=data['area_ha']
struct=data['structure_count']
```

```python
plt.figure(figsize=(8,5))
plt.bar(area, struct,color='blue',alpha=0.7)
plt.title('Area vs Structure Count')
plt.xlabel("Area")
plt.ylabel("Structure Count")
plt.show()
```

```python
sns.set_context("talk")
sns.lineplot(x="age_of_is",y="structure_count",data=data)
plt.show()
```
